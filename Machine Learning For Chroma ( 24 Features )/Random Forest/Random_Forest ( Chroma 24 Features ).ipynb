```python
import numpy as np
import os
import pandas as pd
Choma_folder = "D:\Swarming project buzz1\Features Extracted Data\Chroma_Features"

Noneswarming_train = np.load(os.path.join(Choma_folder, 'Train', 'Noneswarming_train.npy'))
Swarming_train = np.load(os.path.join(Choma_folder, 'Train', 'Swarming_train.npy'))

Noneswarming_test = np.load(os.path.join(Choma_folder, 'Test', 'Noneswarming_test.npy'))
Swarming_test = np.load(os.path.join(Choma_folder, 'Test', 'Swarming_test.npy'))

Noneswarming_val = np.load(os.path.join(Choma_folder, 'Val', 'Noneswarming_val.npy'))
Swarming_val = np.load(os.path.join(Choma_folder, 'Val', 'Swarming_val.npy'))
```

```python
Noneswarming_train_df = pd.DataFrame(np.load(os.path.join(Choma_folder, 'Train', 'Noneswarming_train.npy')))
Swarming_train_df = pd.DataFrame(np.load(os.path.join(Choma_folder, 'Train', 'Swarming_train.npy')))

Noneswarming_test_df = pd.DataFrame(np.load(os.path.join(Choma_folder, 'Test', 'Noneswarming_test.npy')))
Swarming_test_df = pd.DataFrame(np.load(os.path.join(Choma_folder, 'Test', 'Swarming_test.npy')))

Noneswarming_val_df = pd.DataFrame(np.load(os.path.join(Choma_folder, 'Val', 'Noneswarming_val.npy')))
Swarming_val_df = pd.DataFrame(np.load(os.path.join(Choma_folder, 'Val', 'Swarming_val.npy')))
```

```python
X_train = pd.concat([Noneswarming_train_df, Swarming_train_df], ignore_index=True)
X_test = pd.concat([Noneswarming_test_df, Swarming_test_df], ignore_index=True)
X_val = pd.concat([Noneswarming_val_df, Swarming_val_df], ignore_index=True)
```

```python
X_train
```

```python
# Create labels
y_train = pd.Series([0] * Noneswarming_train_df.shape[0] + [1] * Swarming_train_df.shape[0], name='label')
y_test = pd.Series([0] * Noneswarming_test_df.shape[0] + [1] * Swarming_test_df.shape[0], name='label')
y_val = pd.Series([0] * Noneswarming_val_df.shape[0] + [1] * Swarming_val_df.shape[0], name='label')
```

```python
# Importing necessary libraries.
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler and scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
# Initialize the Random Forest model
random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
import time
t1 = time.time()
# Train the model
random_forest_model.fit(X_train_scaled, y_train)

# Make predictions on the validation set
y_test_pred = random_forest_model.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_test_pred)
print(f"Testing Accuracy: {accuracy:.2f}")

# Additional evaluation metrics
print("\nClassification Report:")
print(classification_report(y_test, y_test_pred))
t2 = time.time()
```

```python
print(f"time elapsed: {t2-t1} seconds or {int((t2-t1)//60)} minutes and {int((t2-t1)-(t2-t1)//60*60)} seconds")
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Make predictions on the test set
y_test_prob = random_forest_model.predict_proba(X_test_scaled)[:, 1]

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_test_prob)
roc_auc = roc_auc_score(y_test, y_test_prob)

# Plot the ROC curve
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")

# Save the plot as an image file (e.g., PNG)
plt.savefig('ROC of Random Forest (24 chroma features).png')

# Display the plot
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Make predictions on the test set
y_test_pred = random_forest_model.predict(X_test_scaled)

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_test_pred)

# Plot confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Non-Swarming', 'Swarming'],
            yticklabels=['Non-Swarming', 'Swarming'])
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title('Confusion Matrix')
plt.savefig('Confusion Matrix of Randomfoest (24 Chroma features).png')
plt.show()
```
